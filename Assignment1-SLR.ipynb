```python
#import pandas library and assign to pd for data manipulation and analysis
import pandas as pd
```

```python
#The csv file named "Salary_Data.csv" and stores its content in a pandas dataframe then assigned to the variable "dataset" 
dataset = pd.read_csv("Salary_Data.csv")
```

```python
#The variable "dataset" displays the contents of the pandas dataframe that contains the data read from the csv file named "Salary_Data.csv"

dataset
```

```python
#The independent varaible loaded the column "YearsExperience" from the "dataset" dataframe and creating a bew dataframe with only that column
independent = dataset[["YearsExperience"]]
```

```python
#call the action that only the column named "YearsExperience" as the previous code
independent
```

```python
#the dependent varaible is assigned the column "Salary" from the dataframe and creating a new dataframe with only that column named "Salary"
dependent = dataset[["Salary"]]
```

```python
dependent
```

```python
#The function "train_test_split" from sklearn.model_selection module which is used for splitting the dataset into training and testing for machine learning models
from sklearn.model_selection import train_test_split

#The independent and dependent data into training and testing subsets with 70% of the data allocated for training and 30% for testing. The random_state parameter is set to 0 for reroducibility. The subsets are assigned to varaibles X_train,X_test,y_train and y_test.
X_train,X_test,Y_train,y_test=train_test_split(independent, dependent, test_size=0.30, random_state=0)
```

```python
#The "Linearregression" imported from sklearn.linear_model which is used for create a linear regression model for predictive analysis.
from sklearn.linear_model import LinearRegression
#The LinearRegression class assigned to the variable "regressor" for make predictions using a linear regression model.
regressor = LinearRegression()

#X_train as the independent variable, and Y_train as the dependent variable, allowing the model to learn and find the best fit for the given training data.
regressor.fit(X_train,Y_train)
```

```python
#The variable "weight" means the coefficients(weights) of the linear regression model, which were learned during the fitting process
weight = regressor.coef_
weight
```

```python
#The "bias" means the intercept(bias) term of the linear regression model, which means the predicted value when all independent variables are zero
bais = regressor.intercept_
bais
```

```python
#The variable "y_pred" contains the predicted values of the dependent variable (Salary) based on the linear regression model using the independent variable values (YearsExperience) from the test dataset.
y_pred = regressor.predict(X_test)
```

```python
# the "r2_score" function imported from the "sklearn.metrics" module, which is used to calculate the R-squared score for evaluating the performance of a regression model.
from sklearn.metrics import r2_score
#The variable "r_score" means the R-squared score, which is a measure of how well the predicted values (y_pred) from the linear regression model match the actual values (y_test) of the dependent variable (Salary) in the test dataset.
r_score = r2_score(y_test, y_pred)
```

```python
r_score
```

```python
#imports the "pickle" module and defines the variable "filename" as "finalized_model_linear.sav", which is the filename that will be used to save the trained linear regression model using pickle
import pickle 
filename = "finalized_model_linear.sav"
```

```python
#The code saves the trained linear regression model (regressor) to a file named "finalized_model_linear.sav" using pickle serialization, which allows the model to be stored and loaded later for future use.
pickle.dump(regressor, open(filename, 'wb'))
```

```python
#The code loads the trained linear regression model from the file "finalized_model_linear.sav" using pickle and assigns it to the variable "loaded_model". The model can now be used for making predictions or further analysis.
loaded_model = pickle.load(open("finalized_model_linear.sav", 'rb'))
#The "result" contains the predicted value of the dependent variable (Salary) based on the loaded linear regression model ("loaded_model"), using the independent variable value of 13 (YearsExperience) as input.
result = loaded_model.predict([[13]])
```
